```python
import pandas as pd
import numpy as np
import datetime as dt
```

```python
profile= pd.read_csv('profile.csv') # Demographic data, e.g.age, salary, and gender. 
transcipt= pd.read_csv('transcript.csv') # Transactional data and Offer Stages
portfolio= pd.read_csv('portfolio.csv') # Promotional offers
print(profile.shape,transcipt.shape,portfolio.shape)
```

## Customer Profile

```python
#Demographic
profile=profile.iloc[:,1:]
profile.head(10)
```

```python
#profile.info()
profile['gender']=profile['gender'].fillna("Unknown")
profile['gender']=profile['gender'].apply(lambda x: 'Female' if x=='F' else("Male" if x=='M' else "Undefined"))
profile['income']=profile['income'].fillna(profile['income'].mean())
profile['became_member_on']=profile['became_member_on'].apply(lambda x: dt.datetime.strptime(str(x),'%Y%m%d').date())
#profile.info()
```

```python
def split_value(x):
    ss=len(x.split(','))
    if ss==1:
        return x.split(':')[0],x.split(':')[1],'',''
    elif ss==2:
        aa=[]
        for xx in range(0,2): 
            for yy in range(0,2):
                aa.append(x.split(',')[xx].split(':')[yy])
        return aa[0],aa[1],aa[2],aa[3]
        
```

```python
transcipt['c1']=transcipt['value'].apply(lambda x:split_value(x)[0])
transcipt['c2']=transcipt['value'].apply(lambda x:split_value(x)[1])
transcipt['c3']=transcipt['value'].apply(lambda x:split_value(x)[2])
transcipt['c4']=transcipt['value'].apply(lambda x:split_value(x)[3])

transcipt['c1']=transcipt['c1'].apply(lambda x: x[2:-1])
transcipt['c2']=transcipt['c2'].apply(lambda x: x[2:-2] if x[1]=="'" else x[:-1])
transcipt['c3']=transcipt['c3'].apply(lambda x: x[2:-1] if len(x)>0 else "")
transcipt['c4']=transcipt['c4'].apply(lambda x: x[:-1] if len(x)>0 else "")
```

```python
transcipt.head(5)
```

```python
transcipt['offer_id']=transcipt.apply(lambda x: x['c2'] if x['c1']=='offer id' else '',axis=1)
transcipt['offer_id']=transcipt.apply(lambda x: x['c2'] if x['c1']=='offer_id' else x['offer_id'],axis=1)
transcipt['amount']=transcipt.apply(lambda x: x['c2'] if x['c1']=='amount' else '',axis=1)
transcipt['reward']=transcipt.apply(lambda x: x['c4'] if x['c3']=='reward' else '',axis=1)
```

```python
transcipt.columns=['xxx', 'person', 'event', 'value', 'time', 'c1', 'c2', 'c3','c4', 'offer_id', 'amount', 'reward']
```

```python
transcipt=transcipt.drop(columns=['c1','c2','c3','c4'])
transcipt=transcipt.merge(portfolio,how='left',left_on='offer_id',right_on='id')
```

```python
transcipt["channels"]=transcipt["channels"].fillna("None")
transcipt["web"] = transcipt["channels"].map(lambda x: 1 if "web" in x else 0)
transcipt["email"] = transcipt["channels"].map(lambda x: 1 if "email" in x else 0)
transcipt["mobile"] = transcipt["channels"].map(lambda x: 1 if "mobile" in x else 0)
transcipt["social"] = transcipt["channels"].map(lambda x: 1 if "social" in x else 0)
```
